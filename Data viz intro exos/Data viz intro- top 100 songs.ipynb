```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
dataO = pd.read_csv("Data/Top 100 most Streamed.csv")
dataO.head()
```

```python
correlations = dataO.corr()
correlations
sns.heatmap(correlations, cmap='Purples').set_title("Correlations")
```

```python
dataO.hist()
plt.show()
```

```python
#generes before editing
sns.countplot(y='top genre', data=dataO).set_title("Genres avant dition")
plt.show()
plt.savefig("gen avant dition.png")
```

```python
data = dataO

data['top genre'].replace(['cali rap', 'dfw rap', 'chicago rap', 'emo rap', 'melodic rap', 'pop rap', 'gangster rap'], 'rap', inplace=True)
data['top genre'].replace(['dance pop','canadian pop','australian pop','folk-pop','acoustic pop','danish pop', 'boy band', 'pop dance'], 'pop', inplace=True)
data['top genre'].replace(['north carolina hip hop', 'conscious hip hop', 'canadian hip hop', 'lgbtq+ hip hop', 'detroit hip hop'], 'hip hop', inplace=True)
data['top genre'].replace(['classic rock', 'alternative rock', 'beatlesque', 'modern rock'], 'rock', inplace=True)
data['top genre'].replace(['electropop', 'electro house'], 'electro', inplace=True)
data['top genre'].replace(['canadian contemporary r&b', 'neo soul', 'irish singer-songwriter', 'brostep', 'latin', 'australian dance'], 'other', inplace=True)

sns.countplot(y='top genre', data=data, order = data['top genre'].value_counts().index).set_title("Genres updated")
```

```python
#Popularity vs Danceability 
sns.kdeplot(x="popularity", y="danceability",data=data, shade=True).set_title("Popularity vs Danceability")
```

```python
#Loudness vs energy
rel = sns.relplot(data=data, y='loudness.dB', x='energy')
rel.fig.suptitle("Loudness vs Energy", x=0.6, y=1.1, fontsize=17)
```

```python
#popularity vs year
sns.boxplot(x='year', y='popularity', data=data).set_title("Popularity vs anne d'origine")
plt.xticks(rotation = 90)
```
